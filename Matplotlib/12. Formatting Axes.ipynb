## Formatting Axes

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head()
```

```python
male = df[df['Customer Gender']=='M'].Revenue.resample('M').count()
female = df[df['Customer Gender']=='F'].Revenue.resample('M').count()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(df.resample('M').sum().index,male,female,lw=3)
ax.legend(['Male','Female'])
ax.set_title('Customer Gender',pad=20)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.set_ylim(0,4000)
plt.show()
```
