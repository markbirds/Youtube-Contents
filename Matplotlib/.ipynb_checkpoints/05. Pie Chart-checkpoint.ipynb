## Pie Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head(5)
```

```python
revCountries = df.groupby('Country').Profit.sum()

plt.pie(revCountries,labels=revCountries.index,autopct='%0.2f%%',explode=[0,0,0.2,0,0,0])
plt.title('Profit Percentage from different countries')
plt.axis('equal')
plt.show()
```
