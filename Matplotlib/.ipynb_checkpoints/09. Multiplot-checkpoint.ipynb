## Multiplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1.2,1.2])
ax3 = fig.add_axes([0.04,0.65,0.5,0.5])
ax2 = fig.add_axes([0.07,0.65,0.4,0.5])
ax3.set_xticks([])
ax3.set_yticks([])

ax1Data = df[df['Country']=='United States'].Revenue.resample('M').sum()
ax2Data = df[df['Country']=='United States'].groupby('Product Category').Revenue.sum()

ax1.stackplot(ax1Data.index,ax1Data)
ax2.pie(ax2Data,labels=ax2Data.index)

ax1.set_title('Monthly Revenue from Unites States')
plt.show()
```
