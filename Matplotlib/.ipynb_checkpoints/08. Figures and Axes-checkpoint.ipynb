## Figures and Axes

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head()
```

```python
fig = plt.figure(facecolor='#ff9f36')
ax = fig.add_axes([0,0,1,1],fc='#ffe8cf')
dataframe = pd.crosstab(index=df['Age Group'],columns=df['Product Category'],values=df['Order Quantity'],aggfunc='sum').reindex(['Youth (<25)','Young Adults (25-34)','Adults (35-64)','Seniors (64+)'])
xpos = np.arange(len(dataframe))

ax.set_xticks(xpos)
ax.set_xticklabels(dataframe.index)
ax.bar(xpos-0.2,dataframe.Accessories,width=0.2)
ax.bar(xpos,dataframe.Bikes,width=0.2)
ax.bar(xpos+0.2,dataframe.Clothing,width=0.2)
ax.legend(dataframe.columns)
ax.set_xlabel('Age Group')
ax.set_ylabel('Order Quantity')
ax.set_title('Order Quantity from different Products based on Age Group')
plt.show()
```
