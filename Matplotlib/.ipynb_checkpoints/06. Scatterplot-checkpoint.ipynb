## Scatterplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
dataframe = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
dataframe.head(5)
```

```python
x = np.random.randint(0,113036,size=(200))
```

```python
x
```

```python
age = dataframe.iloc[x]['Customer Age']
order = dataframe.iloc[x]['Order Quantity']
```

```python
plt.scatter(age,order)
plt.title('Relationship between Customer Age and Order Quantity')
plt.xlabel('Customer Age')
plt.ylabel('Order Quantity')
plt.show()
```
