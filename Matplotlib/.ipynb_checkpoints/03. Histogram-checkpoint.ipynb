## Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('data/sampleData2.csv',parse_dates=['Date'],index_col='Date')
```

```python
df.head(5)
```

```python
x = df['Customer Age']

plt.hist(x,bins=np.arange(10,100,10),rwidth=0.8)
plt.title('Customer Age')
plt.xlabel('Age')
plt.show()
```
