## Scatter plot using relplot(),regplot() and lmplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### relplot()

```python
sns.relplot(x="total_bill", y="tip",hue='sex',data=tips)
plt.show()
```

### regplot()

```python
sns.regplot(x="total_bill", y="tip",data=tips)
plt.show()
```

### lmplot()

```python
# regplot with FacetGrid

sns.lmplot(x="total_bill", y="tip",row='time',col='sex',hue='sex',data=tips)
plt.show()
```
