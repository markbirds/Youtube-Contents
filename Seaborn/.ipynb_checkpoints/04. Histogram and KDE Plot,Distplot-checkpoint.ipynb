## Histogram and KDE Plot,Distplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
iris = sns.load_dataset('iris')
```

```python
titanic.head()
```

```python
iris.head()
```

### Histogram

```python
sns.displot(titanic['age'].dropna(),bins=8,palette='magma')
plt.show()
```

### KDE Plot,Distplot

```python
# kde using displot

x1 = iris[iris['species']=='setosa']['petal_length']
x2 = iris[iris['species']=='setosa']['petal_width']

sns.displot([x1,x2],kind='kde',color='b',label=None)
plt.show()
```

```python
setosa_petal_length = iris[iris['species']=='setosa']['petal_length']
setosa_petal_width = iris[iris['species']=='setosa']['petal_width']
versicolor_petal_length = iris[iris['species']=='versicolor']['petal_length']
versicolor_petal_width = iris[iris['species']=='versicolor']['petal_width']
virginica_petal_length = iris[iris['species']=='virginica']['petal_length']
virginica_petal_width = iris[iris['species']=='virginica']['petal_width']

sns.kdeplot(setosa_petal_length,color='b',shade=True,label='setosa petal length')
sns.kdeplot(setosa_petal_width,color='r',shade=True,label='setosa petal width')
plt.legend()
plt.show()
```

```python
sns.kdeplot(x=setosa_petal_length,y=setosa_petal_width,label='setosa petal',cmap='Blues')
sns.kdeplot(x=versicolor_petal_length,y=versicolor_petal_width,label='versicolor petal',cmap='Oranges')
sns.kdeplot(x=virginica_petal_length,y=virginica_petal_width,label='virginica petal',cmap='Greens')
plt.legend()
plt.title('Iris species')
plt.show()
```
