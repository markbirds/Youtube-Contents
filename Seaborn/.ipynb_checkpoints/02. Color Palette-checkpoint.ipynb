```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
plt.figure(figsize=(12,8))
sns.relplot(x="age",y="fare",data=titanic,hue='sex',palette="inferno")
plt.show()
```

```python
sns.relplot(x="age",y="fare",data=titanic,col="class",hue='sex',col_wrap=2,palette="Reds")
plt.show()
```
