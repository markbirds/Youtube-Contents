## FacetGrid and PairGrid

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### FacetGrid

```python
fgrid = sns.FacetGrid(tips, col = "day",hue='sex',col_wrap=2)
fgrid.map(sns.scatterplot,'total_bill','tip')
plt.show()
```

```python
fgrid = sns.FacetGrid(tips, col = "day",col_wrap=2)
fgrid.map(sns.kdeplot,'tip')
plt.show()
```

### PairGrid

```python
pgrid = sns.PairGrid(tips)
pgrid.map_diag(sns.kdeplot,lw=3,color='red')
pgrid.map_upper(sns.scatterplot,color='orange')
pgrid.map_lower(sns.kdeplot,cmap='inferno')
plt.show()
```
