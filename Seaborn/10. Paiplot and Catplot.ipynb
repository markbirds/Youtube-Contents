## Paiplot and Catplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Paiplot

```python
sns.pairplot(tips,hue='time')
plt.show()
```

### Catplot

```python
sns.catplot(x='sex',y='tip',col='day',col_wrap=2,data=tips,kind='swarm')
plt.show()
```

```python
sns.catplot(x='sex',y='tip',col='day',col_wrap=2,data=tips,kind='box')
plt.show()
```
