## Styling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplotlib to seaborn

x = np.arange(1,21)
y1 = x*2
y2 = x*3

print(x)
print(y1)
print(y2)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
sns.set()
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# set_style()
# darkgrid, whitegrid, dark, white, ticks

# set_context()
# paper, notebook, talk, poster

plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style("darkgrid") 
sns.set_context("talk")
sns.despine()
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style("whitegrid") 
sns.set_context("paper")
sns.despine()
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style("dark") 
sns.set_context("notebook")
sns.despine()
plt.show()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style("white") 
sns.set_context("poster")
sns.despine()
plt.show()
```

```python
sns.axes_style()
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
sns.set_style({'grid.color': '.9'})
sns.set_context(None)
plt.show()
```
