## Plotting categorical data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.stripplot(x = "sex", y = "tip", data = tips, jitter=True)
plt.show()
```

```python
sns.swarmplot(x = "sex", y = "tip", data = tips)
plt.show()
```
